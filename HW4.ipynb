### 2.1
Изучите распределение домов от наличия вида на набережную.
Постройте график.

```python
df_w = df['waterfront'].value_counts()
df_w
```

```python
plt.pie(df_w, autopct='%1.1f%%', labels=['Не имеют вида\nна набережную', 'Имеют вид\nна набережную'])
plt.title('Распределение домов в зависимости от наличия вида на набережную');
```

```python
print(f'Средняя стоимость жилья с видом на набережную {df[df["waterfront"] == 1]["price"].mean()}\nСредняя стоимость жилья без вида на набережную {df[df["waterfront"] == 0]["price"].mean()}')
```

У подавляющего большинства недвижимости нет вида на набережную. 

Средняя стоимость недвижимости с видом на набережную примерно в 3 раза выше. 

### 2.2 Изучите распределение этажей домов

```python
df_fl = df['floors'].value_counts()
df_fl
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x = df_fl.index, y = df_fl)
plt.title('Распределение домов в зависимости от количества этажей')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.xticks(rotation=45);
```

### 2.3 Изучите распределение состояния домов

```python
df_con = df['condition'].value_counts()
df_con
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x = df_con.index, y = df_con)
plt.title('Распределение состояния домов')
plt.xlabel('Состояние дома')
plt.ylabel('Количество домов');
```

Большая часть домов находится в среднем состоянии (3 класс).

## Задача 3

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg') 
sns.jointplot(x=df['price'], y=df['sqft_above'], kind='reg')
sns.jointplot(x=df['price'], y=df['grade'], kind='reg')
sns.jointplot(x=df['price'], y=df['bedrooms'], kind='reg')
sns.jointplot(x=df['price'], y=df['bathrooms'], kind='reg')
sns.jointplot(x=df['price'], y=df['yr_built'], kind='reg');
```

Из диаграмм следует, что наибольшее воздействие на цену недвижимости оказывают: 
- площадь жилых помещений;
- общая площадь;
- уровень качества и дизайна строительства;
- количество санузлов. 


Меньшее влияние имеют: 
- год постройки;
- количество спален.
